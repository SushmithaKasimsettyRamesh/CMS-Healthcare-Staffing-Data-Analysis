```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and prepare the datasets
# Assuming you've already loaded the data as pbj_df and penalties_df

# 2. Join the datasets
merged_df = pd.merge(
    pbj_df,
    penalties_df,
    left_on='PROVNUM',
    right_on='CMS Certification Number (CCN)',
    how='inner'  # Only keep facilities that appear in both datasets
)

# 3. Calculate key metrics for each facility
facility_metrics = merged_df.groupby('PROVNUM').agg({
    'PROVNAME': 'first',
    'STATE': 'first',
    'MDScensus': 'mean',
    'Fine Amount': 'sum',
    'WorkDate': 'count'  # Number of days with data
}).reset_index()

# Calculate total and contractor hours by staff type
for staff_type in ['RN', 'LPN', 'CNA']:
    facility_metrics[f'Total_{staff_type}_Hours'] = merged_df.groupby('PROVNUM')[f'Hrs_{staff_type}'].sum().values
    facility_metrics[f'Contractor_{staff_type}_Hours'] = merged_df.groupby('PROVNUM')[f'Hrs_{staff_type}_ctr'].sum().values

# Calculate total hours and contractor percentage
facility_metrics['Total_Hours'] = (
    facility_metrics['Total_RN_Hours'] +
    facility_metrics['Total_LPN_Hours'] +
    facility_metrics['Total_CNA_Hours']
)

facility_metrics['Total_Contractor_Hours'] = (
    facility_metrics['Contractor_RN_Hours'] +
    facility_metrics['Contractor_LPN_Hours'] +
    facility_metrics['Contractor_CNA_Hours']
)

facility_metrics['Contractor_Percentage'] = (
    facility_metrics['Total_Contractor_Hours'] / facility_metrics['Total_Hours'] * 100
)

# 4. Identify high-opportunity facilities
facility_metrics['Opportunity_Score'] = (
    facility_metrics['Fine Amount'] / facility_metrics['Contractor_Percentage'].replace(0, 0.1)
)

# Create opportunity tiers
facility_metrics['Opportunity_Tier'] = pd.qcut(
    facility_metrics['Opportunity_Score'],
    q=5,
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)

# 5. Create visualizations

# Visualization 1: Penalty Amount vs Contractor Usage
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=facility_metrics,
    x='Contractor_Percentage',
    y='Fine Amount',
    hue='Opportunity_Tier',
    size='Total_Hours',
    sizes=(20, 200),
    alpha=0.7
)
plt.title('Penalty Amount vs. Contractor Usage')
plt.xlabel('Contractor Percentage')
plt.ylabel('Total Penalty Amount ($)')
plt.axvline(x=10, color='red', linestyle='--', label='10% Contractor Threshold')
plt.legend(title='Opportunity Tier')
plt.savefig('penalty_vs_contractor.png')
plt.close()

# Visualization 2: Top 10 States by Opportunity Score
top_states = facility_metrics.groupby('STATE')['Opportunity_Score'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_states.index, y=top_states.values)
plt.title('Top 10 States by Sales Opportunity')
plt.xlabel('State')
plt.ylabel('Opportunity Score (Sum)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_states_opportunity.png')
plt.close()

# Visualization 3: Opportunity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(
    data=facility_metrics,
    x='Opportunity_Score',
    hue='Opportunity_Tier',
    multiple='stack',
    bins=20
)
plt.title('Distribution of Sales Opportunities')
plt.xlabel('Opportunity Score')
plt.ylabel('Number of Facilities')
plt.tight_layout()
plt.savefig('opportunity_distribution.png')
plt.close()

# 6. Export target list for sales team
high_opportunity = facility_metrics[
    (facility_metrics['Opportunity_Tier'].isin(['High', 'Very High'])) &
    (facility_metrics['Contractor_Percentage'] < 10) &
    (facility_metrics['Fine Amount'] > 10000)
].sort_values('Opportunity_Score', ascending=False)

# Export to CSV
high_opportunity.to_csv('high_opportunity_targets.csv', index=False)

# 7. Create summary dashboard
print("Summary of High Opportunity Targets:")
print(f"Total facilities identified: {len(high_opportunity)}")
print(f"Total potential penalty savings: ${high_opportunity['Fine Amount'].sum():,.2f}")
print(f"Average contractor percentage: {high_opportunity['Contractor_Percentage'].mean():.2f}%")
print(f"Top 5 states by target count:")
print(high_opportunity['STATE'].value_counts().head(5))
```
